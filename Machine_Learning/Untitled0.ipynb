```python
# Part 2: Load the dataset into a Pandas dataframe.
data = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/netflix_titles.csv')
```

```python
# Next, display a technical summary of the data to determine the data types of each column, and which columns have missing data.
data.info()
```

```python
tv_shows = data[data['type'] == 'TV Show']
TV_Y7_shows = tv_shows[tv_shows['rating'] == 'TV-Y7']
print(f'The number of TV-Y7 shows is: {len(TV_Y7_shows)}')
```

```python
# Further filter your dataset so it only contains TV Shows released between the years 2000 and 2009 inclusive.
TV_shows_2000s = tv_shows[(tv_shows['release_year'] >= 2000) & (tv_shows['release_year'] <= 2009)]
print(f'The number of tv shows released between the years 2000 and 2009 is: {len(TV_shows_2000s)} ')
```

```python
# Part 1: Enter your code below to import Pandas according to the
# conventional method. Then load the dataset into a Pandas dataframe.
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/cereal.csv')
```

```python
# After importing the dataset into a pandas data frame, determine the median amount of protein in cereal brands manufactured by Kelloggs. (mfr code "K")
kelloggs_df = df[df['mfr'] == 'K']
median_protein_kelloggs = kelloggs_df['protein'].median()
print(median_protein_kelloggs)
```

```python
# In order to comply with new government regulations, all cereals must now come with a Healthiness rating. This rating is calculated based on the following formula:
# healthiness = (protein + fiber) / sugar
# Create a new healthiness column populated with values based on the above formula.
df['healthiness'] = (df['protein'] + df['fiber']) / df['sugars']

# Then, determine the median healthiness value for only General Mills cereals (mfr = "G"), rounded to two decimal places.
general_mills_df = df[df['mfr'] == 'G']
median_healthiness_general_mills = general_mills_df['healthiness'].median()
rounded_median_healthiness_general_mills = round(median_healthiness_general_mills, 2)
print(rounded_median_healthiness_general_mills)
```

```python
# Part 1: Enter your code below to import Pandas according to the
# conventional method. Then load the dataset into a Pandas dataframe.
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/titanic.csv')
```

```python
# Create the 'NameGroup' column
df['NameGroup'] = df['Name'].apply(lambda x: x.split(',')[0][0].lower())

# Count how many passengers have a NameGroup value of 'k'
k_count = df[df['NameGroup'] == 'k'].shape[0]

print(f"Number of passengers with NameGroup 'k': {k_count}")
```
